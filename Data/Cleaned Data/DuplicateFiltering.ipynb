Players that appear in the train data are removed from the test set. Strictly focuses the young players from 2020-2025

```python
nba_test_clustered = pd.read_csv("NBA_Test_Clustered.csv")
nba_train = pd.read_csv("NBA_Train_Clustered.csv")

print(f" NBA test data: {len(nba_test_clustered)} players")
print(f" Train test data: {len(nba_train)} players")

testNames = nba_test_clustered["PLAYER"].unique()
trainNames = nba_train["PLAYER"].unique()

uniqueNames=[]
for i in testNames:
  if i not in trainNames:
    uniqueNames.append(i)
print(len(uniqueNames))

filtered_test=nba_test_clustered[nba_test_clustered['PLAYER'].isin(uniqueNames)]
#print(len(filtered_df))
filtered_test.to_csv("FilteredTestData.csv",index=False)

```
